```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
import os
for dirname, _, filenames in os.walk('ParticipantData_BTPC/'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('ParticipantData_BTPC/Train.csv')
data.head()
```

```python
data = data.drop(['session_id','session_number','client_agent','device_details'], axis=1)
data.head()
```

```python
dates = data['date'].values
dat = pd.to_datetime(dates)

data['Month'] = dat.strftime("%m")
data['Year'] = dat.strftime("%Y")
data.head()
```

```python
d = pd.to_datetime(data['date'])
data['dayofWeek'] = d.dt.day_name()
data.head()
```

```python
data['dayofWeek'] = data['dayofWeek'].map({'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,
                                           'Friday':5, 'Saturday':6,'Sunday':7})
data.head()
```

```python
data = data.drop(['date'], axis=1)
data.head()
```

```python
x = data.drop(['time_spent'], axis=1)
y = data['time_spent']
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
y_pred
```

```python
target = y.values
```

```python
model.get_params(deep=True)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(target, y_pred)
```
